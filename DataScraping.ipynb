```python
#importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
#Pull information from the url
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
#Specify what data we need
table = soup.find('table', class_ = 'wikitable sortable')
```

```python
titles = soup.find_all('th')[1:7]
titles
```

```python
#List comprehension to remove tags
table_titles = [title.text.strip() for title in titles]
```

```python
print(table_titles)
```

```python
#Putting column titles into data frame
df = pd.DataFrame(columns= table_titles)
df
```

```python
#Getting data from the table
column_data = table.find_all('tr')
```

```python
#Adding every line of a table to a data frame with for loop using length and loc
for row in column_data[2:]:
  row_data = row.find_all('td')
  ind_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] = ind_row_data[:6]
```

```python
df
```

```python
#Saving data frame as csv file
df.to_csv('/content/sample_data/companies.csv', index=False)
```
